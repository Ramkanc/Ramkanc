```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
col1 = np.random.randint(1,20,10)
col2 = np.random.randint(1,30,10)
print(col1)
print(col2)
```

```python
mean1 = np.mean(col1)
mean2 = np.mean(col2)
print(mean1)
print(mean2)
```

```python
plt.scatter(np.arange(10), col1, marker='o', linestyle='-')
plt.scatter(np.arange(10), col2, marker='x', linestyle='-')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
